```python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Apo_dihedral.csv")
array_chi1 = df['A:268-chi1']
array_chi2 = df['A:268-chi2']
#array_chi3 = df['A:268-chi3']
N = 30

gs = gridspec.GridSpec(2, 6)
ax1 = plt.subplot(gs[:1, :2], polar=True)
ax2 = plt.subplot(gs[:1, 2:])


sns.distplot(array_chi1, bins=N, ax=ax2, color = "grey")
sns.distplot(array_chi2, bins=N, ax=ax2, color = "black")

radii_1, theta_1 = np.histogram(array_chi1, bins=N,density=True)
radii_2, theta_2 = np.histogram(array_chi2, bins=N,density=True)

ax1.set_xticklabels(['{}째'.format(x) for x in [0, 45, 90, 135, 180, -135, -90, -45]])
#ax2.set_xticklabels(['{}째'.format(x) for x in [-180, -135, -90, -45, 0, 45, 90, 135, 180]])
_=plt.xlim([-200, 200])
width = (2*np.pi) / N


bars = ax1.bar(np.pi * theta_1[1:] / 180., radii_1, width=width, color= "grey")
bars = ax1.bar(np.pi * theta_2[1:] / 180., radii_2, width=width, color= "black")


plt.tight_layout()
plt.savefig('radial-apo_Tyr268.pdf', dpi=600)
```

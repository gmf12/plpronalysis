```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

```python
PCA_full=pd.read_csv("//Users//glauciomonteiroferreira//OneDrive//COVID-19//PLP-pro//PCA//Plpro_PCA_merged_proj.csv")
Apo=PCA_full.loc[1:10025:1]
lig_916=PCA_full.loc[10026:20050:1]
lig_917=PCA_full.loc[20050:30075:1]
lig_854=PCA_full.loc[30075:40014:1]



plot = sns.jointplot(x="PC_1", y="PC_2", data= Apo, color = "silver", kind="kde", fill=True)


plot.ax_marg_x.set_xlim(-40, 40)
plot.ax_marg_y.set_ylim(-100, 60)

flatui = ["#4E8AA3", "#67C7F0","#F0CB5B","#582EA3","#894FF0","#A36258","#F0D83A","#7DF0D4","#1A5CA3","#F044BD","#59A39B"]


fig1 = plt.gcf()
plt.rcParams['axes.labelsize'] = 24
plt.rcParams['axes.titlesize'] = 24
plt.xlabel("PC1 (29.3%)")
plt.ylabel("PC2 (17.7%)")
plt.show()
plt.draw()
fig1.savefig('Apo.pdf', dpi=600)
```
